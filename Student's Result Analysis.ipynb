```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Dropping unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# Transforming data

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct", "5-10")
df.head()
```

# Analysis

### Gender distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x ="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

#### From the above chart we have analyzed that the number of females is more than number of males

### Impact of Parent's Edcuation

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's score")
plt.show()
```

#### From the above chart we can conclude that parent's education has a positive impact on their scores 

### Impact of Parent's Marital status

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital status and Student's score")
plt.show()
```

#### From the above chart we can conclude that parent's marital status has negligible impact on their scores

### Distribution of Ethnic Groups

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E" ]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic groups")
plt.show()
```

```python
ax = sns.countplot(data = df, x ="EthnicGroup")
ax.bar_label(ax.containers[0])
```

### Impact of practising Sports

```python
gb2 = df.groupby("PracticeSport").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb2)
```

```python
sns.heatmap(gb2, annot = True)
plt.title("Relationship between practising Sports and Student's score")
plt.show()
```

#### From the above chart we can conclude that practising sports regularly has a positive impact on their scores 

### Impact of Weekly Study Hours

```python
gb3 = df.groupby("WklyStudyHours").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb3)
```

```python
sns.heatmap(gb3, annot = True)
plt.title("Relationship between Weekly Study Hours and Student's score")
plt.show()
```

#### From the above chart we can conclude that studying >5 hours wekly has a positive impact on their scores 

#### Finally,we can conclude the parent's higher education, practising sports regularly, studing >5 hours every week has positive impact on the student's performance
